### Промпт пользователя:
### Найди локальные экстремумы функции $f(x, y) = x^2 + y^2 - xy - x - y$ при условии, что $x + y = 3$. Используй метод множителей Лагранжа и подробно распиши каждый этап вычислений.

### ответ ассистента
:Для нахождения условного экстремума воспользуемся методом множителей Лагранжа.Шаг 1. Составление функции Лагранжа запишем уравнение связи в виде $g(x, y) = x + y - 3 = 0$.Функция Лагранжа имеет вид:$$L(x, y, \lambda) = f(x, y) + \lambda \cdot g(x, y) = x^2 + y^2 - xy - x - y + \lambda(x + y - 3)$$Шаг 2. Поиск стационарных точекНайдем частные производные функции Лагранжа по переменным $x, y$ и $\lambda$, и приравняем их к нулю, чтобы получить систему уравнений:$$\begin{cases}
\frac{\partial L}{\partial x} = 2x - y - 1 + \lambda = 0 \\
\frac{\partial L}{\partial y} = 2y - x - 1 + \lambda = 0 \\
\frac{\partial L}{\partial \lambda} = x + y - 3 = 0
\end{cases}$$Вычтем второе уравнение из первого:$$(2x - y - 1 + \lambda) - (2y - x - 1 + \lambda) = 0$$$$3x - 3y = 0 \implies x = y$$Подставим $x = y$ в третье уравнение (уравнение связи):$$x + x - 3 = 0 \implies 2x = 3 \implies x = 1.5$$Следовательно, $y = 1.5$.Найдем значение множителя Лагранжа $\lambda$:$$2(1.5) - 1.5 - 1 + \lambda = 0 \implies 3 - 2.5 + \lambda = 0 \implies \lambda = -0.5$$Мы нашли одну стационарную точку: $M(1.5; 1.5)$ при $\lambda = -0.5$.Шаг 3. Проверка достаточных условий экстремумаЧтобы определить характер экстремума, исследуем знак второго дифференциала функции Лагранжа $d^2L$ в стационарной точке.Найдем вторые частные производные:$$\frac{\partial^2 L}{\partial x^2} = 2$$$$\frac{\partial^2 L}{\partial y^2} = 2$$$$\frac{\partial^2 L}{\partial x \partial y} = -1$$Запишем второй дифференциал:$$d^2L = \frac{\partial^2 L}{\partial x^2}dx^2 + 2\frac{\partial^2 L}{\partial x \partial y}dxdy + \frac{\partial^2 L}{\partial y^2}dy^2 = 2dx^2 - 2dxdy + 2dy^2$$Из уравнения связи $x + y = 3$ найдем связь между дифференциалами:$$d(x + y) = d(3) \implies dx + dy = 0 \implies dy = -dx$$Подставим $dy = -dx$ в выражение для $d^2L$:$$d^2L = 2dx^2 - 2dx(-dx) + 2(-dx)^2 = 2dx^2 + 2dx^2 + 2dx^2 = 6dx^2$$Так как $d^2L = 6dx^2 > 0$ при любых $dx \neq 0$, квадратичная форма является положительно определенной. Следовательно, в точке $M(1.5; 1.5)$ функция достигает условного минимума.Шаг 4. Вычисление значения функции в точке экстремума$$f_{min}(1.5, 1.5) = (1.5)^2 + (1.5)^2 - (1.5 \cdot 1.5) - 1.5 - 1.5 = 2.25 + 2.25 - 2.25 - 3 = -0.75$$Ответ: Функция имеет условный минимум в точке $(1.5; 1.5)$. Значение функции в этой точке равно $-0.75$.
