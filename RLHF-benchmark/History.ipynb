Основной причиной конфликта стало решение советского руководства тайно разместить ядерное оружие на острове Куба для защиты режима Фиделя Кастро от возможного американского вторжения.

* СССР перебросил на Кубу баллистические ракеты средней дальности. Когда американские самолеты-разведчики обнаружили стартовые площадки, президент США Джон Кеннеди объявил о военно-морской блокаде (карантине) острова, требуя немедленного демонтажа вооружений.

* Кризис разрешился путем дипломатических переговоров между Кеннеди и Никитой Хрущевым. Советский Союз согласился убрать свои ракеты с Кубы, а Соединенные Штаты дали публичное обещание не осуществлять военное вторжение на остров и не свергать кубинское правительство.
