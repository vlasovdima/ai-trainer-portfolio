### ответ ассистента:

Переход к Новой экономической политике (НЭП), провозглашенный на X съезде РКП(б) в марте 1921 года, был вынужденной мерой, обусловленной комплексом причин.

### Три ключевых экономических фактора:

1. Коллапс сельского хозяйства и последствия продразверстки. Крестьяне, у которых насильственно изымали все излишки продукции, потеряли мотивацию к труду. Это привело к массовому сокращению посевных площадей (почти на 25% по сравнению с дореволюционным временем) и падению урожайности. Результатом стал катастрофический дефицит продовольствия и начинающийся голод в ряде губерний.
2. Остановка промышленности и транспортный кризис. Из-за нехватки сырья и топлива (в частности, угля и нефти) большинство крупных предприятий остановилось. Железнодорожная сеть была разрушена за годы Гражданской войны, что сделало невозможным нормальный товарообмен между городом и деревней. Выплавка чугуна и добыча угля упали в несколько раз по сравнению с 1913 годом.
3. Деклассирование пролетариата и кризис системы распределения. Из-за остановки заводов и голода в городах начался массовый отток рабочих в деревню в поисках пропитания. Натурализация оплаты труда и карточная система распределения полностью исчерпали себя, так как государству было нечего распределять.


Два политических фактора:

1. Масштабные крестьянские восстания («Малая Гражданская война»). Недовольство продразверсткой переросло в открытое вооруженное сопротивление. Наиболее крупным стало Тамбовское восстание (Антоновщина) в 1920–1921 годах, а также мятежи в Поволжье, Сибири и на Урале. Крестьянство, бывшее союзником большевиков, превратилось в прямую угрозу их власти.

2. Кронштадтское восстание (март 1921 г.). Вооруженное выступление гарнизона крепости Кронштадт и экипажей Балтийского флота стало решающим фактором. Матросы, ранее считавшиеся «красой и гордостью революции», выдвинули политические требования: «Власть Советам, а не партиям!» и экономические: свободу торговли для крестьян. Восстание показало, что большевики теряют поддержку даже в своей главной социальной базе — армии и флоте.
