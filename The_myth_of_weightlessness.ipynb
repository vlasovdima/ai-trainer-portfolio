Аудит фактологии. Физика (Миф о невесомости)

Промпт пользователя:

"Объясни, почему космонавты на Международной космической станции (МКС) испытывают состояние невесомости. Правда ли, что на низкой околоземной орбите отсутствует гравитация?"

---
Сгенерированный ответ модели (Содержит ошибки):

Космонавты на МКС находятся в состоянии невесомости, потому что станция располагается в зоне отсутствия гравитации (нулевой гравитации). МКС вращается на высоте около 400 километров над поверхностью Земли. На таком огромном расстоянии гравитационное притяжение нашей планеты ослабевает настолько, что становится практически равным нулю. Именно из-за отсутствия силы тяжести, притягивающей их вниз, космонавты и предметы могут свободно парить внутри станции.

Аудит:

Статус: Ответ содержит критические фактические ошибки (научные галлюцинации).

Найденные ошибки и их разбор:Миф о "нулевой гравитации" на высоте 400 км.


Ошибка: Утверждение, что гравитация Земли на высоте МКС равна нулю.

Факт: Сила тяжести убывает обратно пропорционально квадрату расстояния от центра Земли: $g = G \frac{M}{(R+h)^2}$. На высоте $h = 400$ км ускорение свободного падения составляет около $8.7 \text{ м/с}^2$, что всего на 10-15% меньше, чем на поверхности Земли ($9.8 \text{ м/с}^2$). Гравитация там присутствует в полной мере.Неверное объяснение природы невесомости.

Ошибка: Модель утверждает, что объекты парят из-за отсутствия силы, тянущей их вниз.

Факт: МКС и все объекты внутри нее находятся в состоянии непрерывного свободного падения на Землю. Однако станция движется с первой космической скоростью (около $7.9 \text{ км/с}$) по касательной к орбите. Из-за этой огромной горизонтальной скорости поверхность Земли "уходит" из-под станции с той же скоростью, с какой станция падает. Возникает эффект центробежной силы (в неинерциальной системе отсчета), который компенсирует силу тяжести, и вес (сила воздействия на опору) становится равен нулю, хотя гравитация продолжает действовать.
